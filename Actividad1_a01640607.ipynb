##Filtros para variables cualitativas

```python
#Filtramos de la variable NOMBRE el registro LETICIA RAMIREZ para que
#solo nos mostrara la informaci√≥n de leticia
filtro1=df[df["NOMBRE"] == "LETICIA RAMIREZ HERNANDEZ"]
filtro1
```

```python
#Filas no consecutivas, filtramos la fila 3 la fila 245 y la fila 1425
#solo nos motro esas tres filas el filto
filtro2=df.iloc[ [3, 245, 1425] , : ]
filtro2
```

```python
#Filtro por columnas consecutivas
filtro3=df.iloc[ : , 1:3]
filtro3
```

```python
#Filtro por cabecera, solo mostrando los primeros dos registros
filtro4=df.head(2)
filtro4
```

```python
#Filtro por cola es para mostrar los ultimos registros, yo pedi que
#me mostrara los ultimos tres
filtro5=df.tail(3)
filtro5
```

##Filtros para variables cuantitativas

```python
#Filtro por comparacion para que mem mostrara todoso los registros encontrados
#en la variable "CANT" mayores a 50
filtro6=df[df["CANT"] > 50]
filtro6
```

```python
#Para convertir los registros a float tuve que quitar los str primero en la variable
#'COSTO_INITARIO'
#df["COSTO_UNITARIO"] = df["COSTO_UNITARIO"].str.replace("$", "").str.replace(",", "").astype(float)
```

```python
#Mostramos todos los registros encontrados en "costo_unitario" mayores a 100
filtro7=df[df["COSTO_UNITARIO"] > 100]
filtro7
```

```python
#Filtramos los registros menores a 100 en la columna "COSTO_UNITARIO" por nombre

filtro8=df[(df["COSTO_UNITARIO"] < 100) & (df["NOMBRE"] == "AARON NU√ëO CORDOVA")]
filtro8
```

```python
#Filtramos los registros menores a 100 en la columna "COSTO_UNITARIO" y que nos mostrara
#a todos menos a "AARON NU√ëO CORDOVA"
filtro9=df[~(df["COSTO_UNITARIO"] <100) & ~(df["NOMBRE"] == "AARON NU√ëO CORDOVA")]
filtro9
```
